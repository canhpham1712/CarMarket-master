```python
# Ki·ªÉm tra gi√° tr·ªã null trong c·ªôt 'mileage'
print("üîç Ki·ªÉm tra gi√° tr·ªã null trong c·ªôt 'mileage':")
print("\n" + "="*60)

if 'df_merged' in locals():
    # Ki·ªÉm tra tr√™n file merged
    null_count = df_merged['mileage'].isnull().sum()
    total_count = len(df_merged)
    non_null_count = total_count - null_count
    
    print(f"\nüìä Tr√™n file MERGED:")
    print(f"   - T·ªïng s·ªë d√≤ng: {total_count}")
    print(f"   - C√≥ gi√° tr·ªã: {non_null_count} ({non_null_count/total_count*100:.1f}%)")
    print(f"   - Null/Empty: {null_count} ({null_count/total_count*100:.1f}%)")
    
    # Ki·ªÉm tra theo ngu·ªìn
    print(f"\nüìä Ph√¢n t√≠ch theo ngu·ªìn:")
    print(df_merged.groupby('source')['mileage'].apply(lambda x: x.isnull().sum()).to_frame('Null Count'))
    print("\n   T·ª∑ l·ªá null theo ngu·ªìn:")
    print(df_merged.groupby('source')['mileage'].apply(lambda x: f"{x.isnull().sum()}/{len(x)} ({x.isnull().sum()/len(x)*100:.1f}%)"))
    
    # Hi·ªÉn th·ªã m·ªôt s·ªë d√≤ng c√≥ null
    if null_count > 0:
        print(f"\n‚ö†Ô∏è M·ªôt s·ªë d√≤ng c√≥ 'mileage' = null (hi·ªÉn th·ªã 10 d√≤ng ƒë·∫ßu):")
        null_mileage = df_merged[df_merged['mileage'].isnull()]
        print(null_mileage[['ad_id', 'make', 'model', 'mileage', 'title', 'source']].head(10))
else:
    # Ki·ªÉm tra tr√™n t·ª´ng file ri√™ng
    print("\nüìä Tr√™n file BONBANH:")
    null_bonbanh = df_bonbanh['mileage'].isnull().sum()
    print(f"   - T·ªïng s·ªë d√≤ng: {len(df_bonbanh)}")
    print(f"   - C√≥ gi√° tr·ªã: {len(df_bonbanh) - null_bonbanh}")
    print(f"   - Null/Empty: {null_bonbanh} ({null_bonbanh/len(df_bonbanh)*100:.1f}%)")
    
    print("\nüìä Tr√™n file OTO:")
    null_oto = df_oto['mileage'].isnull().sum()
    print(f"   - T·ªïng s·ªë d√≤ng: {len(df_oto)}")
    print(f"   - C√≥ gi√° tr·ªã: {len(df_oto) - null_oto}")
    print(f"   - Null/Empty: {null_oto} ({null_oto/len(df_oto)*100:.1f}%)")
    
    # Hi·ªÉn th·ªã m·ªôt s·ªë d√≤ng c√≥ null
    if null_bonbanh > 0:
        print(f"\n‚ö†Ô∏è BONBANH - M·ªôt s·ªë d√≤ng c√≥ 'mileage' = null:")
        print(df_bonbanh[df_bonbanh['mileage'].isnull()][['ad_id', 'make', 'model', 'mileage', 'title']].head(5))
    
    if null_oto > 0:
        print(f"\n‚ö†Ô∏è OTO - M·ªôt s·ªë d√≤ng c√≥ 'mileage' = null:")
        print(df_oto[df_oto['mileage'].isnull()][['ad_id', 'make', 'model', 'mileage', 'title']].head(5))
```

```python
# Ki·ªÉm tra c√°c gi√° tr·ªã v√† format trong c·ªôt 'mileage'
print("üîç Ph√¢n t√≠ch chi ti·∫øt c·ªôt 'mileage':")
print("\n" + "="*60)

if 'df_merged' in locals():
    # T·ªïng quan
    print("\nüìä T·ªîNG QUAN:")
    print(f"   - T·ªïng s·ªë d√≤ng: {len(df_merged)}")
    print(f"   - C√≥ gi√° tr·ªã: {df_merged['mileage'].notna().sum()}")
    print(f"   - Null: {df_merged['mileage'].isnull().sum()}")
    
    # L·∫•y c√°c gi√° tr·ªã kh√¥ng null
    mileage_not_null = df_merged[df_merged['mileage'].notna()]['mileage']
    
    print(f"\nüìã C√ÅC GI√Å TR·ªä UNIQUE (hi·ªÉn th·ªã 50 gi√° tr·ªã ƒë·∫ßu):")
    unique_values = mileage_not_null.unique()
    print(f"   - T·ªïng s·ªë gi√° tr·ªã unique: {len(unique_values)}")
    print(f"\n   Danh s√°ch:")
    for i, val in enumerate(sorted(unique_values)[:50], 1):
        count = (mileage_not_null == val).sum()
        print(f"   {i:2d}. '{val}' ({count} l·∫ßn)")
    
    # Ph√¢n t√≠ch format
    print(f"\nüîç PH√ÇN T√çCH FORMAT:")
    
    # Ki·ªÉm tra c√°c pattern ph·ªï bi·∫øn
    import re
    
    patterns = {
        'C√≥ "km"': r'km',
        'Ch·ªâ s·ªë': r'^\d+$',
        'S·ªë + km': r'^\d+\s*km$',
        'S·ªë + d·∫•u ph·∫©y + km': r'^\d{1,3}(?:,\d{3})+\s*km$',
        'S·ªë + d·∫•u ch·∫•m + km': r'^\d{1,3}(?:\.\d{3})+\s*km$',
        'C√≥ "ngh√¨n"': r'ngh√¨n',
        'C√≥ "v·∫°n"': r'v·∫°n',
    }
    
    for pattern_name, pattern in patterns.items():
        count = mileage_not_null.str.contains(pattern, case=False, na=False).sum()
        print(f"   - {pattern_name}: {count} ({count/len(mileage_not_null)*100:.1f}%)")
    
    # Xem m·ªôt s·ªë v√≠ d·ª• c·ª• th·ªÉ
    print(f"\nüìù M·ªòT S·ªê V√ç D·ª§ C·ª§ TH·ªÇ:")
    sample_mileage = mileage_not_null.head(20)
    for i, val in enumerate(sample_mileage, 1):
        print(f"   {i:2d}. '{val}'")
    
    # Th·ªëng k√™ theo ngu·ªìn
    print(f"\nüìä TH·ªêNG K√ä THEO NGU·ªíN:")
    for source in df_merged['source'].unique():
        source_df = df_merged[df_merged['source'] == source]
        source_mileage = source_df[source_df['mileage'].notna()]['mileage']
        print(f"\n   {source.upper()}:")
        print(f"   - C√≥ gi√° tr·ªã: {len(source_mileage)}")
        print(f"   - Unique: {len(source_mileage.unique())}")
        print(f"   - M·ªôt s·ªë v√≠ d·ª•:")
        for val in source_mileage.head(5):
            print(f"     ‚Ä¢ '{val}'")
    
    # Ki·ªÉm tra c√°c gi√° tr·ªã b·∫•t th∆∞·ªùng ho·∫∑c kh√¥ng h·ª£p l·ªá
    print(f"\n‚ö†Ô∏è KI·ªÇM TRA GI√Å TR·ªä B·∫§T TH∆Ø·ªúNG:")
    
    # T√¨m c√°c gi√° tr·ªã kh√¥ng c√≥ s·ªë
    no_number = mileage_not_null[~mileage_not_null.str.contains(r'\d', na=False)]
    if len(no_number) > 0:
        print(f"   - Kh√¥ng ch·ª©a s·ªë: {len(no_number)} gi√° tr·ªã")
        print(f"     V√≠ d·ª•: {no_number.unique()[:5].tolist()}")
    
    # T√¨m c√°c gi√° tr·ªã c√≥ s·ªë nh∆∞ng format l·∫°
    has_number = mileage_not_null[mileage_not_null.str.contains(r'\d', na=False)]
    unusual = has_number[~has_number.str.contains(r'^\d+.*km', case=False, na=False)]
    if len(unusual) > 0:
        print(f"   - Format kh√¥ng chu·∫©n (c√≥ s·ªë nh∆∞ng kh√¥ng ph·∫£i 's·ªë km'): {len(unusual)} gi√° tr·ªã")
        print(f"     V√≠ d·ª•: {unusual.unique()[:10].tolist()}")
```

```python
# Chu·∫©n h√≥a c·ªôt mileage: b·ªè "km" v√† chuy·ªÉn th√†nh s·ªë
print("üîÑ Chu·∫©n h√≥a c·ªôt 'mileage':")
print("\n" + "="*60)

if 'df_merged' in locals():
    import re
    
    def normalize_mileage(mileage_value):
        """
        Chu·∫©n h√≥a mileage:
        - B·ªè "km", "kilomet", etc.
        - X·ª≠ l√Ω d·∫•u ph·∫©y, d·∫•u ch·∫•m
        - X·ª≠ l√Ω "ngh√¨n", "k", "v·∫°n"
        - Chuy·ªÉn th√†nh s·ªë nguy√™n
        """
        if pd.isna(mileage_value):
            return None
        
        mileage_str = str(mileage_value).strip()
        
        # B·ªè "km", "kilomet", "kilometer" (case-insensitive)
        mileage_str = re.sub(r'\s*(km|kilomet|kilometer)\s*', '', mileage_str, flags=re.IGNORECASE)
        
        # X·ª≠ l√Ω "ngh√¨n" ho·∫∑c "k" (v√≠ d·ª•: "19 ngh√¨n", "19k")
        if 'ngh√¨n' in mileage_str.lower() or re.search(r'(\d+)\s*k\b', mileage_str.lower()):
            match = re.search(r'(\d+(?:\.\d+)?)', mileage_str)
            if match:
                num = float(match.group(1))
                return int(num * 1000)
        
        # X·ª≠ l√Ω "v·∫°n" (v√≠ d·ª•: "1.5 v·∫°n" = 15000)
        if 'v·∫°n' in mileage_str.lower():
            match = re.search(r'(\d+(?:\.\d+)?)', mileage_str)
            if match:
                num = float(match.group(1))
                return int(num * 10000)
        
        # B·ªè d·∫•u ph·∫©y v√† d·∫•u ch·∫•m (format s·ªë: 100,000 ho·∫∑c 100.000)
        mileage_str = mileage_str.replace(',', '').replace('.', '')
        
        # Extract s·ªë
        match = re.search(r'(\d+)', mileage_str)
        if match:
            return int(match.group(1))
        
        return None
    
    # Xem m·ªôt s·ªë v√≠ d·ª• tr∆∞·ªõc khi chu·∫©n h√≥a
    print("\nüìã M·ªôt s·ªë v√≠ d·ª• TR∆Ø·ªöC khi chu·∫©n h√≥a:")
    sample_before = df_merged[df_merged['mileage'].notna()]['mileage'].head(10)
    for val in sample_before:
        normalized = normalize_mileage(val)
        print(f"   '{val}' -> {normalized}")
    
    # Chu·∫©n h√≥a
    print(f"\nüîÑ ƒêang chu·∫©n h√≥a...")
    df_merged['mileage'] = df_merged['mileage'].apply(normalize_mileage)
    
    # Xem k·∫øt qu·∫£
    print(f"\n‚úÖ ƒê√£ chu·∫©n h√≥a xong!")
    print(f"\nüìä K·∫øt qu·∫£:")
    print(f"   - Ki·ªÉu d·ªØ li·ªáu: {df_merged['mileage'].dtype}")
    print(f"   - Null: {df_merged['mileage'].isnull().sum()}")
    print(f"   - C√≥ gi√° tr·ªã: {df_merged['mileage'].notna().sum()}")
    
    print(f"\nüìã M·ªôt s·ªë v√≠ d·ª• SAU khi chu·∫©n h√≥a:")
    sample_after = df_merged[df_merged['mileage'].notna()]['mileage'].head(10)
    for val in sample_after:
        print(f"   {val}")
    
    # Th·ªëng k√™
    print(f"\nüìà Th·ªëng k√™:")
    print(f"   - Min: {df_merged['mileage'].min()}")
    print(f"   - Max: {df_merged['mileage'].max()}")
    print(f"   - Mean: {df_merged['mileage'].mean():.0f}")
    print(f"   - Median: {df_merged['mileage'].median():.0f}")
```

```python
# Fill mileage null theo ph∆∞∆°ng ph√°p ƒë∆∞·ª£c ƒë·ªÅ xu·∫•t (c√≥ x·ª≠ l√Ω ƒë·∫∑c bi·ªát cho nƒÉm 2025)
print("üîÑ FILL MILEAGE NULL THEO PH∆Ø∆†NG PH√ÅP ƒê∆Ø·ª¢C ƒê·ªÄ XU·∫§T")
print("\n" + "="*60)

if 'df_merged' in locals():
    null_before = df_merged['mileage'].isnull().sum()
    print(f"\nüìä Tr∆∞·ªõc khi fill: {null_before} null")
    
    # Ph√¢n t√≠ch null theo year
    null_by_year = df_merged[df_merged['mileage'].isnull()]['year'].value_counts()
    print(f"\nüìä Ph√¢n b·ªë null theo year:")
    print(null_by_year.head(5))
    
    # T√≠nh c√°c median ƒë·ªÉ s·ª≠ d·ª•ng
    print(f"\nüîπ T√≠nh to√°n c√°c median...")
    
    # Median t·ªïng th·ªÉ
    median_overall = df_merged['mileage'].median()
    print(f"   - Median t·ªïng th·ªÉ: {median_overall:.0f} km")
    
    # Median theo Make + Model (cho nƒÉm 2025)
    median_by_make_model = df_merged.groupby(['make', 'model'])['mileage'].median()
    print(f"   - Median theo Make+Model: ƒë√£ t√≠nh cho {len(median_by_make_model)} nh√≥m")
    
    # Median theo Make + Model + Year (cho c√°c nƒÉm kh√°c)
    median_by_group = df_merged.groupby(['make', 'model', 'year'])['mileage'].median()
    print(f"   - Median theo Make+Model+Year: ƒë√£ t√≠nh cho {len(median_by_group)} nh√≥m")
    
    # H√†m fill th√¥ng minh
    def fill_mileage_smart(row):
        if pd.notna(row['mileage']):
            return row['mileage']
        
        make = row['make']
        model = row['model']
        year = row['year']
        
        # X·ª≠ l√Ω ƒë·∫∑c bi·ªát cho nƒÉm 2025 (xe m·ªõi)
        if pd.notna(year) and year == 2025:
            # D√πng median theo Make + Model (kh√¥ng theo year) v√¨ xe m·ªõi c√≥ mileage th·∫•p
            key = (make, model)
            if key in median_by_make_model.index:
                median_val = median_by_make_model[key]
                if pd.notna(median_val):
                    return median_val
        
        # C√°c nƒÉm kh√°c: d√πng median theo Make + Model + Year
        if pd.notna(year):
            key = (make, model, year)
            if key in median_by_group.index:
                median_val = median_by_group[key]
                if pd.notna(median_val):
                    return median_val
        
        # Fallback 1: Median theo Make + Model
        key = (make, model)
        if key in median_by_make_model.index:
            median_val = median_by_make_model[key]
            if pd.notna(median_val):
                return median_val
        
        # Fallback 2: Median t·ªïng th·ªÉ
        return median_overall
    
    # Fill mileage
    print(f"\nüîÑ ƒêang fill mileage...")
    mask_null = df_merged['mileage'].isnull()
    
    # Fill t·ª´ng d√≤ng
    filled_count = 0
    for idx in df_merged[mask_null].index:
        old_val = df_merged.loc[idx, 'mileage']
        new_val = fill_mileage_smart(df_merged.loc[idx])
        df_merged.loc[idx, 'mileage'] = new_val
        if pd.isna(old_val) and pd.notna(new_val):
            filled_count += 1
    
    null_after = df_merged['mileage'].isnull().sum()
    
    print(f"‚úÖ Ho√†n th√†nh!")
    print(f"   - Tr∆∞·ªõc: {null_before} null")
    print(f"   - Sau: {null_after} null")
    print(f"   - ƒê√£ fill: {filled_count} gi√° tr·ªã")
    
    # Th·ªëng k√™ theo nƒÉm sau khi fill
    print(f"\nüìä Th·ªëng k√™ sau khi fill:")
    if 'year' in df_merged.columns:
        print(f"\n   Mileage trung b√¨nh theo year (sau fill):")
        avg_by_year = df_merged.groupby('year')['mileage'].mean()
        print(avg_by_year.head(10))
        
        # ƒê·∫∑c bi·ªát ki·ªÉm tra nƒÉm 2025
        year_2025 = df_merged[df_merged['year'] == 2025]
        if len(year_2025) > 0:
            print(f"\n   NƒÉm 2025 (xe m·ªõi):")
            print(f"   - S·ªë l∆∞·ª£ng: {len(year_2025)}")
            print(f"   - Mileage trung b√¨nh: {year_2025['mileage'].mean():.0f} km")
            print(f"   - Mileage median: {year_2025['mileage'].median():.0f} km")
            print(f"   - Min: {year_2025['mileage'].min():.0f} km")
            print(f"   - Max: {year_2025['mileage'].max():.0f} km")
```

```python
# Ki·ªÉm tra t√≠nh ƒë√∫ng ƒë·∫Øn c·ªßa c·ªôt version
print("üîç KI·ªÇM TRA T√çNH ƒê√öNG ƒê·∫ÆN C·ª¶A C·ªòT 'version'")
print("\n" + "="*60)

if 'df_merged' in locals():
    # 1. Th√¥ng tin c∆° b·∫£n
    print("\n1Ô∏è‚É£ TH√îNG TIN C∆† B·∫¢N:")
    print(f"   - Ki·ªÉu d·ªØ li·ªáu: {df_merged['version'].dtype}")
    print(f"   - T·ªïng s·ªë d√≤ng: {len(df_merged)}")
    print(f"   - Null: {df_merged['version'].isnull().sum()} ({df_merged['version'].isnull().sum()/len(df_merged)*100:.1f}%)")
    print(f"   - Kh√¥ng null: {df_merged['version'].notna().sum()} ({df_merged['version'].notna().sum()/len(df_merged)*100:.1f}%)")
    
    # 2. Ki·ªÉm tra c√°c gi√° tr·ªã unique
    print(f"\n2Ô∏è‚É£ C√ÅC GI√Å TR·ªä UNIQUE:")
    version_counts = df_merged['version'].value_counts()
    print(f"   - T·ªïng s·ªë gi√° tr·ªã unique: {len(version_counts)}")
    print(f"   - Top 20 gi√° tr·ªã ph·ªï bi·∫øn nh·∫•t:")
    print(version_counts.head(20))
    
    # 3. Ki·ªÉm tra version c√≥ xu·∫•t hi·ªán trong title kh√¥ng (validation)
    print(f"\n3Ô∏è‚É£ KI·ªÇM TRA VERSION C√ì TRONG TITLE (Validation):")
    
    def check_version_in_title(row):
        """Ki·ªÉm tra xem version c√≥ xu·∫•t hi·ªán trong title kh√¥ng"""
        if pd.isna(row['version']) or pd.isna(row['title']):
            return None
        
        version_str = str(row['version']).strip()
        title_str = str(row['title']).lower()
        
        # Chu·∫©n h√≥a ƒë·ªÉ so s√°nh
        version_normalized = version_str.lower().replace(' ', '').replace('-', '')
        title_normalized = title_str.replace(' ', '').replace('-', '')
        
        # Ki·ªÉm tra version c√≥ trong title kh√¥ng
        if version_normalized in title_normalized:
            return True
        
        # Ki·ªÉm tra m·ªôt ph·∫ßn c·ªßa version (v√≠ d·ª•: "1.5E" trong "1.5E CVT")
        version_parts = version_normalized.split()
        if any(part in title_normalized for part in version_parts if len(part) > 2):
            return True
        
        return False
    
    df_merged['version_in_title'] = df_merged.apply(check_version_in_title, axis=1)
    
    version_with_title = df_merged[df_merged['version'].notna()]['version_in_title']
    match_stats = version_with_title.value_counts()
    
    print(f"   - C√≥ version: {len(version_with_title)} d√≤ng")
    print(f"   - Version c√≥ trong title: {match_stats.get(True, 0)} ({match_stats.get(True, 0)/len(version_with_title)*100:.1f}%)")
    print(f"   - Version kh√¥ng c√≥ trong title: {match_stats.get(False, 0)} ({match_stats.get(False, 0)/len(version_with_title)*100:.1f}%)")
    
    # Hi·ªÉn th·ªã m·ªôt s·ªë v√≠ d·ª• version kh√¥ng c√≥ trong title
    if match_stats.get(False, 0) > 0:
        print(f"\n   ‚ö†Ô∏è M·ªôt s·ªë v√≠ d·ª• version KH√îNG c√≥ trong title:")
        examples = df_merged[(df_merged['version'].notna()) & (df_merged['version_in_title'] == False)][
            ['make', 'model', 'version', 'title']
        ].head(10)
        for idx, row in examples.iterrows():
            print(f"     ‚Ä¢ {row['model']} - Version: '{row['version']}'")
            print(f"       Title: {row['title']}")
    
    # 4. Ki·ªÉm tra format c·ªßa version
    print(f"\n4Ô∏è‚É£ KI·ªÇM TRA FORMAT C·ª¶A VERSION:")
    
    version_not_null = df_merged[df_merged['version'].notna()]['version']
    
    # Ph√¢n lo·∫°i format
    format_patterns = {
        'C√≥ s·ªë + ch·ªØ c√°i (v√≠ d·ª•: 1.5E, 2.5G)': [],
        'Ch·ªâ ch·ªØ c√°i (v√≠ d·ª•: G, E, S)': [],
        'C√≥ CVT/MT/AT': [],
        'C√≥ s·ªë (v√≠ d·ª•: 1.5, 2.0)': [],
        'Format kh√°c': []
    }
    
    import re
    
    for version in version_not_null.unique():
        version_str = str(version).strip()
        
        if re.search(r'\d+\.?\d*[A-Za-z]+', version_str):  # C√≥ s·ªë + ch·ªØ
            format_patterns['C√≥ s·ªë + ch·ªØ c√°i (v√≠ d·ª•: 1.5E, 2.5G)'].append(version_str)
        elif re.search(r'^[A-Za-z\s]+$', version_str) and not re.search(r'\d', version_str):  # Ch·ªâ ch·ªØ
            format_patterns['Ch·ªâ ch·ªØ c√°i (v√≠ d·ª•: G, E, S)'].append(version_str)
        elif re.search(r'CVT|MT|AT|AMT', version_str, re.IGNORECASE):  # C√≥ lo·∫°i h·ªôp s·ªë
            format_patterns['C√≥ CVT/MT/AT'].append(version_str)
        elif re.search(r'^\d+\.?\d*$', version_str):  # Ch·ªâ s·ªë
            format_patterns['C√≥ s·ªë (v√≠ d·ª•: 1.5, 2.0)'].append(version_str)
        else:
            format_patterns['Format kh√°c'].append(version_str)
    
    print(f"   Ph√¢n lo·∫°i format:")
    for category, values in format_patterns.items():
        unique_count = len(set(values))
        total_count = len(values)
        if total_count > 0:
            print(f"\n   {category}:")
            print(f"   - Unique: {unique_count}")
            print(f"   - V√≠ d·ª•: {list(set(values))[:5]}")
    
    # 5. Ki·ªÉm tra version theo model
    print(f"\n5Ô∏è‚É£ KI·ªÇM TRA VERSION THEO MODEL:")
    print(f"   - C√°c model c√≥ nhi·ªÅu version nh·∫•t:")
    version_by_model = df_merged.groupby('model')['version'].agg(['count', 'nunique'])
    version_by_model = version_by_model.sort_values('nunique', ascending=False)
    print(version_by_model.head(10))
    
    # 6. Ki·ªÉm tra c√°c gi√° tr·ªã c√≥ v·∫ª kh√¥ng h·ª£p l·ªá
    print(f"\n6Ô∏è‚É£ KI·ªÇM TRA C√ÅC GI√Å TR·ªä KH√îNG H·ª¢P L·ªÜ:")
    
    suspicious_versions = []
    
    # Ki·ªÉm tra version qu√° d√†i ho·∫∑c qu√° ng·∫Øn
    for version in version_not_null.unique():
        version_str = str(version).strip()
        if len(version_str) > 50:  # Qu√° d√†i
            suspicious_versions.append(('Qu√° d√†i', version_str))
        elif len(version_str) == 0:  # R·ªóng
            suspicious_versions.append(('R·ªóng', version_str))
        elif version_str.lower() in ['null', 'none', 'nan', 'n/a', 'na']:  # Gi√° tr·ªã placeholder
            suspicious_versions.append(('Placeholder', version_str))
    
    if suspicious_versions:
        print(f"   ‚ö†Ô∏è T√¨m th·∫•y c√°c gi√° tr·ªã ƒë√°ng ng·ªù:")
        for reason, value in suspicious_versions:
            count = len(df_merged[df_merged['version'] == value])
            print(f"   - '{value}' ({reason}): {count} d√≤ng")
    else:
        print(f"   ‚úÖ Kh√¥ng t√¨m th·∫•y gi√° tr·ªã ƒë√°ng ng·ªù")
    
    # 7. So s√°nh version gi·ªØa c√°c ngu·ªìn
    print(f"\n7Ô∏è‚É£ SO S√ÅNH VERSION GI·ªÆA C√ÅC NGU·ªíN:")
    for source in df_merged['source'].unique():
        source_df = df_merged[df_merged['source'] == source]
        print(f"\n   {source.upper()}:")
        print(f"   - Null: {source_df['version'].isnull().sum()} ({source_df['version'].isnull().sum()/len(source_df)*100:.1f}%)")
        print(f"   - C√≥ version: {source_df['version'].notna().sum()}")
        print(f"   - Unique versions: {source_df['version'].nunique()}")
        print(f"   - Top 10 version:")
        print(source_df['version'].value_counts().head(10))
    
    # 8. Ki·ªÉm tra version c√≥ t∆∞∆°ng th√≠ch v·ªõi model kh√¥ng
    print(f"\n8Ô∏è‚É£ KI·ªÇM TRA T∆Ø∆†NG TH√çCH VERSION V√Ä MODEL:")
    print(f"   (Ki·ªÉm tra m·ªôt s·ªë model ph·ªï bi·∫øn)")
    
    common_models = ['Vios', 'Innova', 'Corolla', 'Camry', 'Fortuner']
    for model in common_models:
        model_df = df_merged[df_merged['model'] == model]
        if len(model_df) > 0:
            versions = model_df['version'].value_counts().head(5)
            print(f"\n   {model}:")
            print(f"   - Top 5 version:")
            for version, count in versions.items():
                print(f"     ‚Ä¢ {version}: {count} d√≤ng")
```

```python
# Chu·∫©n h√≥a c√°c version t∆∞∆°ng t·ª± nhau
print("üîÑ CHU·∫®N H√ìA C√ÅC VERSION T∆Ø∆†NG T·ª∞")
print("\n" + "="*60)

if 'df_merged' in locals():
    import re
    
    # 1. Chu·∫©n h√≥a kho·∫£ng tr·∫Øng v√† ch·ªØ hoa/th∆∞·ªùng
    print("\n1Ô∏è‚É£ Chu·∫©n h√≥a kho·∫£ng tr·∫Øng v√† ch·ªØ hoa/th∆∞·ªùng...")
    
    def normalize_version_spacing(version):
        """Chu·∫©n h√≥a kho·∫£ng tr·∫Øng trong version"""
        if pd.isna(version):
            return version
        
        version_str = str(version).strip()
        
        # Chu·∫©n h√≥a kho·∫£ng tr·∫Øng: lo·∫°i b·ªè kho·∫£ng tr·∫Øng th·ª´a
        version_str = ' '.join(version_str.split())
        
        # Chu·∫©n h√≥a kho·∫£ng tr·∫Øng quanh d·∫•u ch·∫•m s·ªë (v√≠ d·ª•: "1.5 E" -> "1.5E")
        version_str = re.sub(r'(\d+\.?\d*)\s+([A-Za-z])', r'\1\2', version_str)
        
        # Chu·∫©n h√≥a kho·∫£ng tr·∫Øng quanh CVT/MT/AT
        version_str = re.sub(r'\s+(CVT|MT|AT|AMT)\b', r' \1', version_str, flags=re.IGNORECASE)
        version_str = re.sub(r'\b(CVT|MT|AT|AMT)\s+', r'\1 ', version_str, flags=re.IGNORECASE)
        
        return version_str
    
    # √Åp d·ª•ng chu·∫©n h√≥a
    df_merged['version'] = df_merged['version'].apply(normalize_version_spacing)
    
    print(f"   ‚úÖ ƒê√£ chu·∫©n h√≥a kho·∫£ng tr·∫Øng")
    
    # 2. Chu·∫©n h√≥a c√°c version t∆∞∆°ng t·ª± nhau
    print("\n2Ô∏è‚É£ Chu·∫©n h√≥a c√°c version t∆∞∆°ng t·ª± nhau...")
    
    # Mapping c√°c version t∆∞∆°ng t·ª±
    version_mapping = {
        # Altis variations
        'altis 1.8G AT': 'Altis 1.8G AT',
        'Altis 1.8 G': 'Altis 1.8G',
        'altis 1.8G': 'Altis 1.8G',
        
        # Cross variations
        'Cross 1.8 V': 'Cross 1.8V',
        'Cross 1.8V': 'Cross 1.8V',
        'Cross 1.8 HEV': 'Cross 1.8HEV',
        'Cross 1.8HEV': 'Cross 1.8HEV',
        
        # CVT variations
        '1.5 E CVT': '1.5E CVT',
        '1.5E CVT': '1.5E CVT',
        '1.5 G CVT': '1.5G CVT',
        'G 1.5 CVT': '1.5G CVT',
        
        # MT variations
        '1.5 E MT': '1.5E MT',
        '1.5E MT': '1.5E MT',
        
        # AT variations
        '1.2 G AT': '1.2G AT',
        '1.5G AT': '1.5G AT',
    }
    
    # √Åp d·ª•ng mapping
    count_changed = 0
    for old_version, new_version in version_mapping.items():
        mask = df_merged['version'] == old_version
        if mask.sum() > 0:
            count_changed += mask.sum()
            df_merged.loc[mask, 'version'] = new_version
            print(f"   - '{old_version}' ‚Üí '{new_version}': {mask.sum()} d√≤ng")
    
    print(f"\n   ‚úÖ ƒê√£ chu·∫©n h√≥a {count_changed} d√≤ng")
    
    # 3. Ki·ªÉm tra l·∫°i sau khi chu·∫©n h√≥a
    print("\n3Ô∏è‚É£ Ki·ªÉm tra l·∫°i sau khi chu·∫©n h√≥a:")
    print(f"   - T·ªïng s·ªë unique: {df_merged['version'].nunique()}")
    print(f"\n   Top 20 version sau chu·∫©n h√≥a:")
    print(df_merged['version'].value_counts().head(20))
```

```python
# Ph√¢n t√≠ch t·∫°i sao c√≥ nhi·ªÅu version null
print("üîç PH√ÇN T√çCH T·∫†I SAO C√ì NHI·ªÄU VERSION NULL")
print("\n" + "="*60)

if 'df_merged' in locals():
    # 1. Ki·ªÉm tra c√°c d√≤ng c√≥ version null
    print("\n1Ô∏è‚É£ KI·ªÇM TRA C√ÅC D√íNG C√ì VERSION NULL:")
    null_version = df_merged[df_merged['version'].isnull()]
    print(f"   - T·ªïng s·ªë null: {len(null_version)}")
    print(f"   - Theo ngu·ªìn:")
    print(null_version['source'].value_counts())
    
    # 2. Ki·ªÉm tra title c·ªßa c√°c d√≤ng null c√≥ ch·ª©a version kh√¥ng
    print(f"\n2Ô∏è‚É£ KI·ªÇM TRA TITLE C·ª¶A C√ÅC D√íNG NULL:")
    
    if 'title' in null_version.columns:
        print(f"\n   üìã M·ªôt s·ªë v√≠ d·ª• title c·ªßa c√°c d√≤ng null:")
        examples = null_version[['make', 'model', 'year', 'title', 'source']].head(20)
        
        for idx, row in examples.iterrows():
            print(f"\n   D√≤ng {idx} ({row['source']}):")
            print(f"     Model: {row['model']}, Year: {row['year']}")
            print(f"     Title: {row['title']}")
            
            # Ph√¢n t√≠ch title ƒë·ªÉ t√¨m version
            title_str = str(row['title']).lower()
            
            # T√¨m c√°c pattern version ph·ªï bi·∫øn
            import re
            patterns_found = []
            
            # Pattern: s·ªë + ch·ªØ c√°i (v√≠ d·ª•: 1.5E, 2.0G)
            if re.search(r'\d+\.?\d*[A-Za-z]+', title_str):
                patterns_found.append("C√≥ s·ªë + ch·ªØ c√°i")
            
            # Pattern: CVT, MT, AT
            if re.search(r'\b(CVT|MT|AT|AMT)\b', title_str, re.IGNORECASE):
                patterns_found.append("C√≥ CVT/MT/AT")
            
            # Pattern: G, E, S (ch·ªØ c√°i ƒë∆°n)
            if re.search(r'\b([GESQVX])\b', title_str):
                patterns_found.append("C√≥ ch·ªØ c√°i ƒë∆°n (G/E/S)")
            
            if patterns_found:
                print(f"     ‚ö†Ô∏è T√¨m th·∫•y pattern: {', '.join(patterns_found)}")
            else:
                print(f"     ‚ùå Kh√¥ng t√¨m th·∫•y pattern version r√µ r√†ng")
    
    # 3. Ph√¢n t√≠ch theo model
    print(f"\n3Ô∏è‚É£ PH√ÇN T√çCH NULL THEO MODEL:")
    null_by_model = null_version['model'].value_counts()
    print(null_by_model.head(15))
    
    # 4. Ph√¢n t√≠ch theo ngu·ªìn chi ti·∫øt
    print(f"\n4Ô∏è‚É£ PH√ÇN T√çCH CHI TI·∫æT THEO NGU·ªíN:")
    
    for source in null_version['source'].unique():
        source_null = null_version[null_version['source'] == source]
        print(f"\n   üìä {source.upper()} ({len(source_null)} d√≤ng null):")
        
        # Ph√¢n t√≠ch theo model
        print(f"\n   - Top 10 model c√≥ null:")
        print(source_null['model'].value_counts().head(10))
        
        # Xem m·ªôt s·ªë v√≠ d·ª• title
        print(f"\n   - M·ªôt s·ªë v√≠ d·ª• title:")
        examples = source_null[['model', 'title']].head(10)
        for idx, row in examples.iterrows():
            print(f"     ‚Ä¢ {row['model']}: {row['title'][:80]}...")
    
    # 5. So s√°nh v·ªõi c√°c d√≤ng c√≥ version
    print(f"\n5Ô∏è‚É£ SO S√ÅNH V·ªöI C√ÅC D√íNG C√ì VERSION:")
    
    has_version = df_merged[df_merged['version'].notna()]
    
    print(f"\n   D√≤ng C√ì version:")
    print(f"   - T·ªïng: {len(has_version)}")
    print(f"   - Theo ngu·ªìn:")
    print(has_version['source'].value_counts())
    
    print(f"\n   D√≤ng KH√îNG c√≥ version:")
    print(f"   - T·ªïng: {len(null_version)}")
    print(f"   - Theo ngu·ªìn:")
    print(null_version['source'].value_counts())
    
    # 6. Ki·ªÉm tra xem c√≥ th·ªÉ extract version t·ª´ title kh√¥ng
    print(f"\n6Ô∏è‚É£ TH·ª¨ EXTRACT VERSION T·ª™ TITLE CHO C√ÅC D√íNG NULL:")
    
    def extract_version_from_title(title, model):
        """Th·ª≠ extract version t·ª´ title"""
        if pd.isna(title):
            return None
        
        title_str = str(title).lower()
        model_str = str(model).lower() if pd.notna(model) else ''
        
        # Lo·∫°i b·ªè model v√† make kh·ªèi title ƒë·ªÉ t√¨m version
        title_clean = title_str.replace(model_str, '').strip()
        
        import re
        
        # Pattern 1: S·ªë + ch·ªØ c√°i (v√≠ d·ª•: 1.5E, 2.0G)
        match = re.search(r'(\d+\.?\d*[A-Za-z]+)', title_clean)
        if match:
            return match.group(1).upper()
        
        # Pattern 2: Ch·ªØ c√°i ƒë∆°n + s·ªë (v√≠ d·ª•: E 1.5, G 2.0)
        match = re.search(r'\b([A-Z])\s+(\d+\.?\d*)', title_clean)
        if match:
            return f"{match.group(2)}{match.group(1)}"
        
        # Pattern 3: CVT/MT/AT v·ªõi s·ªë (v√≠ d·ª•: 1.5 CVT)
        match = re.search(r'(\d+\.?\d*)\s*(CVT|MT|AT)', title_clean, re.IGNORECASE)
        if match:
            return f"{match.group(1)}{match.group(2).upper()}"
        
        return None
    
    # Test extract tr√™n m·ªôt s·ªë d√≤ng null
    print(f"\n   Th·ª≠ extract tr√™n 20 d√≤ng null ƒë·∫ßu ti√™n:")
    extracted_count = 0
    for idx, row in null_version.head(20).iterrows():
        extracted = extract_version_from_title(row.get('title', ''), row.get('model', ''))
        if extracted:
            extracted_count += 1
            print(f"     ‚Ä¢ {row['model']}: '{row['title'][:60]}...'")
            print(f"       ‚Üí Extracted: {extracted}")
    
    print(f"\n   ‚úÖ C√≥ th·ªÉ extract ƒë∆∞·ª£c: {extracted_count}/20 d√≤ng")
    
    # 7. ƒê·ªÅ xu·∫•t
    print(f"\n7Ô∏è‚É£ ƒê·ªÄ XU·∫§T:")
    print(f"   - Oto c√≥ {null_version[null_version['source'] == 'oto']['version'].isnull().sum()} d√≤ng null")
    print(f"   - C√≥ th·ªÉ extract version t·ª´ title cho c√°c d√≤ng n√†y")
    print(f"   - C·∫ßn ki·ªÉm tra logic extract trong scraper oto.com.vn")
```

```python
# Extract v√† fill version t·ª´ title cho c√°c d√≤ng null
print("üîÑ EXTRACT V√Ä FILL VERSION T·ª™ TITLE")
print("\n" + "="*60)

if 'df_merged' in locals():
    import re
    
    def extract_version_from_title(title, model, description=None):
        """
        Extract version t·ª´ title v√† description
        X·ª≠ l√Ω nhi·ªÅu pattern kh√°c nhau
        """
        if pd.isna(title):
            return None
        
        title_str = str(title).strip()
        model_str = str(model).lower() if pd.notna(model) else ''
        
        # Lo·∫°i b·ªè model v√† make kh·ªèi title ƒë·ªÉ t√¨m version
        title_clean = title_str.replace(model_str, '').replace('toyota', '').strip()
        
        # N·∫øu c√≥ description, c≈©ng ki·ªÉm tra trong ƒë√≥
        full_text = title_clean
        if description and pd.notna(description):
            full_text += ' ' + str(description)
        
        full_text = full_text.lower()
        
        # Pattern 1: S·ªë + ch·ªØ c√°i + CVT/MT/AT (v√≠ d·ª•: 2.0MT, 1.5E CVT, 2.2 AT)
        # ∆Øu ti√™n pattern n√†y v√¨ ph·ªï bi·∫øn nh·∫•t
        patterns = [
            # Pattern: s·ªë + ch·ªØ c√°i + CVT/MT/AT (v√≠ d·ª•: 2.0MT, 1.5E CVT)
            r'(\d+\.?\d*[A-Za-z]*)\s*(CVT|MT|AT|AMT)',
            # Pattern: s·ªë + CVT/MT/AT (v√≠ d·ª•: 2.0 CVT, 1.5 MT)
            r'(\d+\.?\d*)\s+(CVT|MT|AT|AMT)',
            # Pattern: ch·ªØ c√°i + s·ªë + CVT/MT/AT (v√≠ d·ª•: LE 2.2 AT, G 1.5 CVT)
            r'([A-Za-z]+)\s+(\d+\.?\d*)\s+(CVT|MT|AT|AMT)',
            # Pattern: s·ªë + ch·ªØ c√°i (v√≠ d·ª•: 1.5E, 2.0G, 2.5Q)
            r'(\d+\.?\d*[A-Za-z]+)',
            # Pattern: ch·ªØ c√°i ƒë∆°n + s·ªë (v√≠ d·ª•: E 1.5, G 2.0) - √≠t ph·ªï bi·∫øn h∆°n
            r'\b([A-Z])\s+(\d+\.?\d*)',
        ]
        
        for pattern in patterns:
            match = re.search(pattern, full_text, re.IGNORECASE)
            if match:
                if len(match.groups()) == 2:
                    # Pattern c√≥ 2 groups (v√≠ d·ª•: s·ªë + CVT/MT/AT)
                    if match.group(2).upper() in ['CVT', 'MT', 'AT', 'AMT']:
                        # Tr∆∞·ªùng h·ª£p: s·ªë + CVT/MT/AT
                        version = f"{match.group(1)}{match.group(2).upper()}"
                    else:
                        # Tr∆∞·ªùng h·ª£p: ch·ªØ c√°i + s·ªë
                        version = f"{match.group(2)}{match.group(1).upper()}"
                else:
                    # Pattern c√≥ 1 group
                    version = match.group(1).upper()
                
                # Chu·∫©n h√≥a format
                version = version.strip()
                # ƒê·∫£m b·∫£o CVT/MT/AT ƒë∆∞·ª£c vi·∫øt hoa v√† c√≥ kho·∫£ng tr·∫Øng tr∆∞·ªõc n√≥ n·∫øu c·∫ßn
                version = re.sub(r'(\d+\.?\d*[A-Za-z]*)(CVT|MT|AT|AMT)', r'\1 \2', version, flags=re.IGNORECASE)
                version = re.sub(r'\s+', ' ', version)  # Lo·∫°i b·ªè kho·∫£ng tr·∫Øng th·ª´a
                
                return version
        
        return None
    
    # 1. Ki·ªÉm tra s·ªë l∆∞·ª£ng null tr∆∞·ªõc khi fill
    null_before = df_merged['version'].isnull().sum()
    print(f"\nüìä Tr∆∞·ªõc khi fill:")
    print(f"   - S·ªë d√≤ng null: {null_before}")
    print(f"   - Theo ngu·ªìn:")
    print(df_merged[df_merged['version'].isnull()]['source'].value_counts())
    
    # 2. Extract version t·ª´ title cho c√°c d√≤ng null
    print(f"\nüîÑ ƒêang extract version t·ª´ title...")
    
    null_mask = df_merged['version'].isnull()
    null_rows = df_merged[null_mask]
    
    extracted_count = 0
    failed_count = 0
    
    for idx in null_rows.index:
        row = df_merged.loc[idx]
        title = row.get('title', '')
        model = row.get('model', '')
        description = row.get('description', '')
        
        extracted_version = extract_version_from_title(title, model, description)
        
        if extracted_version:
            df_merged.loc[idx, 'version'] = extracted_version
            extracted_count += 1
        else:
            failed_count += 1
    
    # 3. K·∫øt qu·∫£ sau khi extract
    null_after = df_merged['version'].isnull().sum()
    
    print(f"\n‚úÖ K·∫øt qu·∫£:")
    print(f"   - ƒê√£ extract ƒë∆∞·ª£c: {extracted_count} d√≤ng")
    print(f"   - Kh√¥ng extract ƒë∆∞·ª£c: {failed_count} d√≤ng")
    print(f"   - Null tr∆∞·ªõc: {null_before}")
    print(f"   - Null sau: {null_after}")
    print(f"   - ƒê√£ gi·∫£m: {null_before - null_after} d√≤ng null")
    
    # 4. Xem m·ªôt s·ªë v√≠ d·ª• ƒë√£ extract
    if extracted_count > 0:
        print(f"\nüìã M·ªôt s·ªë v√≠ d·ª• ƒë√£ extract:")
        # L·∫•y c√°c d√≤ng v·ª´a ƒë∆∞·ª£c fill (c√≥ th·ªÉ ki·ªÉm tra b·∫±ng c√°ch so s√°nh)
        examples = df_merged[df_merged['version'].notna()].tail(extracted_count).head(10)
        for idx, row in examples.iterrows():
            print(f"   ‚Ä¢ {row['model']}: '{row['title'][:60]}...'")
            print(f"     ‚Üí Version: {row['version']}")
    
    # 5. Xem c√°c d√≤ng kh√¥ng extract ƒë∆∞·ª£c
    if failed_count > 0:
        print(f"\n‚ö†Ô∏è C√°c d√≤ng kh√¥ng extract ƒë∆∞·ª£c ({failed_count} d√≤ng):")
        still_null = df_merged[df_merged['version'].isnull()]
        examples = still_null[['make', 'model', 'title']].head(10)
        for idx, row in examples.iterrows():
            print(f"   ‚Ä¢ {row['model']}: {row['title']}")
    
    # 6. Th·ªëng k√™ sau khi fill
    print(f"\nüìä Th·ªëng k√™ sau khi fill:")
    print(f"   - T·ªïng s·ªë d√≤ng: {len(df_merged)}")
    print(f"   - C√≥ version: {df_merged['version'].notna().sum()} ({df_merged['version'].notna().sum()/len(df_merged)*100:.1f}%)")
    print(f"   - Null: {df_merged['version'].isnull().sum()} ({df_merged['version'].isnull().sum()/len(df_merged)*100:.1f}%)")
    
    print(f"\n   Theo ngu·ªìn:")
    for source in df_merged['source'].unique():
        source_df = df_merged[df_merged['source'] == source]
        null_source = source_df['version'].isnull().sum()
        print(f"   - {source}: {null_source}/{len(source_df)} null ({null_source/len(source_df)*100:.1f}%)")
```

```python
# Ph√¢n t√≠ch v√† c·∫£i thi·ªán extract version cho c√°c d√≤ng kh√¥ng extract ƒë∆∞·ª£c
print("üîç PH√ÇN T√çCH C√ÅC D√íNG KH√îNG EXTRACT ƒê∆Ø·ª¢C")
print("\n" + "="*60)

if 'df_merged' in locals():
    import re
    
    # L·∫•y c√°c d√≤ng v·∫´n c√≤n null
    still_null = df_merged[df_merged['version'].isnull()]
    print(f"\nüìä C√≤n l·∫°i {len(still_null)} d√≤ng null")
    
    # Ph√¢n t√≠ch c√°c pattern trong c√°c d√≤ng n√†y
    print(f"\n1Ô∏è‚É£ PH√ÇN T√çCH PATTERN TRONG C√ÅC D√íNG NULL:")
    
    patterns_found = {
        'C√≥ XLi/GLi + s·ªë': [],
        'C√≥ GL/G + s·ªë': [],
        'C√≥ TRD/SPORTIVO': [],
        'Ch·ªâ c√≥ nƒÉm': [],
        'Kh√°c': []
    }
    
    for idx, row in still_null.iterrows():
        title = str(row.get('title', '')).lower()
        model = str(row.get('model', '')).lower()
        
        # Lo·∫°i b·ªè model
        text = title.replace(model, '').replace('toyota', '').strip()
        
        # Ph√¢n lo·∫°i
        if re.search(r'(xli|gli)\s+(\d+\.?\d*)', text):
            patterns_found['C√≥ XLi/GLi + s·ªë'].append((idx, row['title']))
        elif re.search(r'\b(gl|g)\s+(\d+\.?\d*)', text):
            patterns_found['C√≥ GL/G + s·ªë'].append((idx, row['title']))
        elif re.search(r'trd|sportivo', text):
            patterns_found['C√≥ TRD/SPORTIVO'].append((idx, row['title']))
        elif re.search(r'^\d{4}|tr∆∞·ªõc\s+\d{4}', text):
            patterns_found['Ch·ªâ c√≥ nƒÉm'].append((idx, row['title']))
        else:
            patterns_found['Kh√°c'].append((idx, row['title']))
    
    for category, items in patterns_found.items():
        if items:
            print(f"\n   {category} ({len(items)} d√≤ng):")
            for idx, title in items[:5]:
                print(f"     ‚Ä¢ {title}")
    
    # 2. C·∫£i thi·ªán h√†m extract
    print(f"\n2Ô∏è‚É£ C·∫¢I THI·ªÜN H√ÄM EXTRACT:")
    
    def extract_version_improved(row):
        """H√†m extract version c·∫£i thi·ªán v·ªõi nhi·ªÅu pattern h∆°n"""
        if pd.notna(row['version']):
            return row['version']
        
        title = str(row.get('title', '')).lower()
        model = str(row.get('model', '')).lower()
        desc = str(row.get('description', '')).lower()
        
        # Lo·∫°i b·ªè model v√† make
        text = title.replace(model, '').replace('toyota', '').strip()
        if desc:
            text += ' ' + desc
        
        # Pattern 1: S·ªë + ch·ªØ c√°i + CVT/MT/AT (∆∞u ti√™n cao nh·∫•t)
        match = re.search(r'(\d+\.?\d*[A-Za-z]*)\s*(CVT|MT|AT|AMT)', text, re.IGNORECASE)
        if match:
            return f"{match.group(1).upper()} {match.group(2).upper()}"
        
        # Pattern 2: S·ªë + CVT/MT/AT
        match = re.search(r'(\d+\.?\d*)\s+(CVT|MT|AT|AMT)', text, re.IGNORECASE)
        if match:
            return f"{match.group(1)} {match.group(2).upper()}"
        
        # Pattern 3: XLi/GLi + s·ªë (v√≠ d·ª•: XLi 2.2, GLi 2.2)
        match = re.search(r'(xli|gli)\s+(\d+\.?\d*)', text, re.IGNORECASE)
        if match:
            return f"{match.group(1).upper()} {match.group(2)}"
        
        # Pattern 4: GL/G + s·ªë (v√≠ d·ª•: GL 2.4, G 2.0)
        match = re.search(r'\b(gl|g)\s+(\d+\.?\d*)', text, re.IGNORECASE)
        if match:
            # Ch·ªâ l·∫•y n·∫øu kh√¥ng ph·∫£i l√† "G" ƒë∆°n gi·∫£n (c√≥ th·ªÉ l√† gearbox)
            if match.group(1).lower() == 'gl' or (match.group(1).lower() == 'g' and len(match.group(2)) > 0):
                return f"{match.group(1).upper()} {match.group(2)}"
        
        # Pattern 5: TRD SPORTIVO ho·∫∑c c√°c version ƒë·∫∑c bi·ªát
        match = re.search(r'(trd\s*sportivo|sportivo|trd)', text, re.IGNORECASE)
        if match:
            return match.group(1).upper()
        
        # Pattern 6: S·ªë + ch·ªØ c√°i (v√≠ d·ª•: 1.5E, 2.0G, 2.5Q)
        match = re.search(r'(\d+\.?\d*[A-Za-z]+)', text)
        if match:
            return match.group(1).upper()
        
        # Pattern 7: Ch·ªØ c√°i ƒë∆°n + s·ªë (v√≠ d·ª•: E 1.5, G 2.0) - √≠t ph·ªï bi·∫øn
        match = re.search(r'\b([A-Z])\s+(\d+\.?\d*)', text)
        if match:
            return f"{match.group(2)}{match.group(1).upper()}"
        
        return None
    
    # 3. Th·ª≠ extract l·∫°i v·ªõi h√†m c·∫£i thi·ªán
    print(f"\n3Ô∏è‚É£ TH·ª¨ EXTRACT L·∫†I V·ªöI H√ÄM C·∫¢I THI·ªÜN:")
    
    null_before = df_merged['version'].isnull().sum()
    print(f"   - Null tr∆∞·ªõc: {null_before}")
    
    # Extract l·∫°i cho c√°c d√≤ng null
    mask_null = df_merged['version'].isnull()
    df_merged.loc[mask_null, 'version'] = df_merged.loc[mask_null].apply(extract_version_improved, axis=1)
    
    null_after = df_merged['version'].isnull().sum()
    newly_extracted = null_before - null_after
    
    print(f"   - Null sau: {null_after}")
    print(f"   - ƒê√£ extract th√™m: {newly_extracted} d√≤ng")
    
    # 4. Xem c√°c v√≠ d·ª• v·ª´a extract ƒë∆∞·ª£c
    if newly_extracted > 0:
        print(f"\nüìã C√°c v√≠ d·ª• v·ª´a extract ƒë∆∞·ª£c:")
        # L·∫•y c√°c d√≤ng v·ª´a ƒë∆∞·ª£c fill (c√≥ th·ªÉ ki·ªÉm tra b·∫±ng c√°ch xem c√°c d√≤ng c√≥ version m·ªõi)
        examples = df_merged[df_merged['version'].notna()].tail(newly_extracted).head(10)
        for idx, row in examples.iterrows():
            print(f"   ‚Ä¢ {row['model']}: {row['title'][:60]}...")
            print(f"     ‚Üí Version: {row['version']}")
    
    # 5. Xem c√°c d√≤ng v·∫´n c√≤n null
    if null_after > 0:
        print(f"\n‚ö†Ô∏è C√°c d√≤ng v·∫´n c√≤n null ({null_after} d√≤ng):")
        still_null = df_merged[df_merged['version'].isnull()]
        examples = still_null[['make', 'model', 'title']].head(10)
        for idx, row in examples.iterrows():
            print(f"   ‚Ä¢ {row['model']}: {row['title']}")
            
            # Ph√¢n t√≠ch t·∫°i sao kh√¥ng extract ƒë∆∞·ª£c
            title = str(row['title']).lower()
            model = str(row['model']).lower()
            text = title.replace(model, '').replace('toyota', '').strip()
            print(f"     ‚Üí Text sau khi lo·∫°i b·ªè model: '{text}'")
    
    # 6. T·ªïng k·∫øt
    print(f"\n‚úÖ T·ªîNG K·∫æT:")
    print(f"   - T·ªïng s·ªë d√≤ng: {len(df_merged)}")
    print(f"   - C√≥ version: {df_merged['version'].notna().sum()} ({df_merged['version'].notna().sum()/len(df_merged)*100:.1f}%)")
    print(f"   - Null: {df_merged['version'].isnull().sum()} ({df_merged['version'].isnull().sum()/len(df_merged)*100:.1f}%)")
    
    print(f"\n   Theo ngu·ªìn:")
    for source in df_merged['source'].unique():
        source_df = df_merged[df_merged['source'] == source]
        null_source = source_df['version'].isnull().sum()
        print(f"   - {source}: {null_source}/{len(source_df)} null ({null_source/len(source_df)*100:.1f}%)")
```

```python
# Ki·ªÉm tra c·ªôt color
print("üîç KI·ªÇM TRA C·ªòT 'color'")
print("\n" + "="*60)

if 'df_merged' in locals():
    # 1. Th√¥ng tin c∆° b·∫£n
    print("\n1Ô∏è‚É£ TH√îNG TIN C∆† B·∫¢N:")
    print(f"   - Ki·ªÉu d·ªØ li·ªáu: {df_merged['color'].dtype}")
    print(f"   - T·ªïng s·ªë d√≤ng: {len(df_merged)}")
    print(f"   - Null: {df_merged['color'].isnull().sum()} ({df_merged['color'].isnull().sum()/len(df_merged)*100:.1f}%)")
    print(f"   - Kh√¥ng null: {df_merged['color'].notna().sum()} ({df_merged['color'].notna().sum()/len(df_merged)*100:.1f}%)")
    
    # 2. C√°c gi√° tr·ªã unique
    print(f"\n2Ô∏è‚É£ C√ÅC GI√Å TR·ªä UNIQUE:")
    color_counts = df_merged['color'].value_counts()
    print(f"   - T·ªïng s·ªë gi√° tr·ªã unique: {len(color_counts)}")
    print(f"\n   Top 30 gi√° tr·ªã ph·ªï bi·∫øn nh·∫•t:")
    print(color_counts.head(30))
    
    # 3. Ph√¢n t√≠ch theo ngu·ªìn
    print(f"\n3Ô∏è‚É£ PH√ÇN T√çCH THEO NGU·ªíN:")
    for source in df_merged['source'].unique():
        source_df = df_merged[df_merged['source'] == source]
        source_color = source_df[source_df['color'].notna()]['color']
        print(f"\n   üìä {source.upper()}:")
        print(f"   - Null: {source_df['color'].isnull().sum()} ({source_df['color'].isnull().sum()/len(source_df)*100:.1f}%)")
        print(f"   - C√≥ m√†u: {len(source_color)}")
        print(f"   - Unique colors: {source_color.nunique()}")
        print(f"   - Top 10 m√†u:")
        print(source_color.value_counts().head(10))
    
    # 4. Ki·ªÉm tra c√°c gi√° tr·ªã c√≥ v·∫ª kh√¥ng h·ª£p l·ªá
    print(f"\n4Ô∏è‚É£ KI·ªÇM TRA C√ÅC GI√Å TR·ªä KH√îNG H·ª¢P L·ªÜ:")
    
    color_not_null = df_merged[df_merged['color'].notna()]['color']
    
    # Ki·ªÉm tra c√°c gi√° tr·ªã ƒë√°ng ng·ªù
    suspicious_colors = []
    
    # Gi√° tr·ªã qu√° d√†i
    long_colors = color_not_null[color_not_null.str.len() > 50]
    if len(long_colors) > 0:
        suspicious_colors.append(('Qu√° d√†i (>50 k√Ω t·ª±)', long_colors.unique()[:5].tolist()))
    
    # Gi√° tr·ªã c√≥ s·ªë (c√≥ th·ªÉ kh√¥ng ph·∫£i m√†u)
    import re
    has_numbers = color_not_null[color_not_null.str.contains(r'\d', na=False)]
    if len(has_numbers) > 0:
        suspicious_colors.append(('C√≥ ch·ª©a s·ªë', has_numbers.unique()[:5].tolist()))
    
    # Gi√° tr·ªã placeholder
    placeholders = color_not_null[color_not_null.str.lower().isin(['null', 'none', 'nan', 'n/a', 'na', ''])]
    if len(placeholders) > 0:
        suspicious_colors.append(('Placeholder', placeholders.unique().tolist()))
    
    if suspicious_colors:
        print(f"   ‚ö†Ô∏è T√¨m th·∫•y c√°c gi√° tr·ªã ƒë√°ng ng·ªù:")
        for reason, values in suspicious_colors:
            print(f"\n   {reason}:")
            for val in values:
                count = len(df_merged[df_merged['color'] == val])
                print(f"     - '{val}': {count} d√≤ng")
    else:
        print(f"   ‚úÖ Kh√¥ng t√¨m th·∫•y gi√° tr·ªã ƒë√°ng ng·ªù")
    
    # 5. Ki·ªÉm tra c√°c m√†u ph·ªï bi·∫øn
    print(f"\n5Ô∏è‚É£ C√ÅC M√ÄU PH·ªî BI·∫æN:")
    
    # Danh s√°ch m√†u xe ph·ªï bi·∫øn ·ªü Vi·ªát Nam
    common_colors_vn = [
        'Tr·∫Øng', 'ƒêen', 'B·∫°c', 'X√°m', 'Xanh', 'ƒê·ªè', 'V√†ng', 'N√¢u',
        'B·∫°c', 'Ghi', 'C√°t', 'V√†ng c√°t', 'N√¢u v√†ng', 'Xanh d∆∞∆°ng',
        'Xanh l√°', 'Xanh ƒëen', 'ƒê·ªè ƒë·∫≠m', 'Tr·∫Øng ng·ªçc', 'ƒêen b√≥ng'
    ]
    
    print(f"   - Top 20 m√†u trong d·ªØ li·ªáu:")
    top_colors = color_counts.head(20)
    for color, count in top_colors.items():
        print(f"     ‚Ä¢ {color}: {count} d√≤ng")
    
    # 6. Ki·ªÉm tra m√†u c√≥ trong title kh√¥ng (validation)
    print(f"\n6Ô∏è‚É£ KI·ªÇM TRA M√ÄU C√ì TRONG TITLE (Validation):")
    
    def check_color_in_title(row):
        """Ki·ªÉm tra xem m√†u c√≥ xu·∫•t hi·ªán trong title kh√¥ng"""
        if pd.isna(row['color']) or pd.isna(row['title']):
            return None
        
        color_str = str(row['color']).strip().lower()
        title_str = str(row['title']).lower()
        
        # Ki·ªÉm tra m√†u c√≥ trong title kh√¥ng
        if color_str in title_str:
            return True
        
        # Ki·ªÉm tra m·ªôt ph·∫ßn c·ªßa m√†u (v√≠ d·ª•: "v√†ng c√°t" c√≥ th·ªÉ c√≥ "v√†ng" ho·∫∑c "c√°t")
        color_parts = color_str.split()
        if any(part in title_str for part in color_parts if len(part) > 2):
            return True
        
        return False
    
    df_merged['color_in_title'] = df_merged.apply(check_color_in_title, axis=1)
    
    color_with_title = df_merged[df_merged['color'].notna()]['color_in_title']
    match_stats = color_with_title.value_counts()
    
    print(f"   - C√≥ m√†u: {len(color_with_title)} d√≤ng")
    print(f"   - M√†u c√≥ trong title: {match_stats.get(True, 0)} ({match_stats.get(True, 0)/len(color_with_title)*100:.1f}%)")
    print(f"   - M√†u kh√¥ng c√≥ trong title: {match_stats.get(False, 0)} ({match_stats.get(False, 0)/len(color_with_title)*100:.1f}%)")
    
    # Hi·ªÉn th·ªã m·ªôt s·ªë v√≠ d·ª• m√†u kh√¥ng c√≥ trong title
    if match_stats.get(False, 0) > 0:
        print(f"\n   ‚ö†Ô∏è M·ªôt s·ªë v√≠ d·ª• m√†u KH√îNG c√≥ trong title:")
        examples = df_merged[(df_merged['color'].notna()) & (df_merged['color_in_title'] == False)][
            ['make', 'model', 'color', 'title']
        ].head(10)
        for idx, row in examples.iterrows():
            print(f"     ‚Ä¢ {row['model']} - M√†u: '{row['color']}'")
            print(f"       Title: {row['title'][:60]}...")
    
    # 7. Ph√¢n t√≠ch null theo model
    print(f"\n7Ô∏è‚É£ PH√ÇN T√çCH NULL THEO MODEL:")
    null_color = df_merged[df_merged['color'].isnull()]
    if len(null_color) > 0:
        print(f"   - Top 10 model c√≥ nhi·ªÅu null nh·∫•t:")
        print(null_color['model'].value_counts().head(10))
    
    # 8. T·ªïng k·∫øt
    print(f"\n8Ô∏è‚É£ T·ªîNG K·∫æT:")
    print(f"   - T·ªïng s·ªë d√≤ng: {len(df_merged)}")
    print(f"   - C√≥ m√†u: {df_merged['color'].notna().sum()} ({df_merged['color'].notna().sum()/len(df_merged)*100:.1f}%)")
    print(f"   - Null: {df_merged['color'].isnull().sum()} ({df_merged['color'].isnull().sum()/len(df_merged)*100:.1f}%)")
    print(f"   - Unique colors: {df_merged['color'].nunique()}")
```

```python
# Extract v√† fill m√†u t·ª´ description
print("üîÑ EXTRACT V√Ä FILL M√ÄU T·ª™ DESCRIPTION")
print("\n" + "="*60)

if 'df_merged' in locals():
    import re
    
    # Danh s√°ch m√†u ph·ªï bi·∫øn trong ti·∫øng Vi·ªát
    color_keywords = {
        'ƒêen': ['ƒëen', 'black'],
        'Tr·∫Øng': ['tr·∫Øng', 'white', 'trang'],
        'B·∫°c': ['b·∫°c', 'silver'],
        'ƒê·ªè': ['ƒë·ªè', 'red', 'do'],
        'C√°t': ['c√°t', 'beige', 'v√†ng c√°t', 'n√¢u v√†ng'],
        'Xanh': ['xanh', 'blue', 'green', 'xanh d∆∞∆°ng', 'xanh l√°'],
        'Ghi': ['ghi', 'gray', 'grey', 'x√°m'],
        'X√°m': ['x√°m', 'gray', 'grey'],
        'V√†ng': ['v√†ng', 'yellow', 'v√†ng c√°t', 'n√¢u v√†ng'],
        'N√¢u': ['n√¢u', 'brown', 'nau'],
        'ƒê·ªìng': ['ƒë·ªìng', 'copper', 'dong'],
        'Cam': ['cam', 'orange'],
        'H·ªìng': ['h·ªìng', 'pink'],
        'T√≠m': ['t√≠m', 'purple'],
        'Kem': ['kem', 'cream']
    }
    
    def extract_color_from_text(text):
        """Extract m√†u t·ª´ text (description ho·∫∑c title)"""
        if pd.isna(text):
            return None
        
        text_lower = str(text).lower()
        
        # T√¨m m√†u theo th·ª© t·ª± ∆∞u ti√™n (m√†u c·ª• th·ªÉ tr∆∞·ªõc, m√†u chung sau)
        # ∆Øu ti√™n c√°c m√†u c·ª• th·ªÉ tr∆∞·ªõc
        priority_colors = ['V√†ng c√°t', 'N√¢u v√†ng', 'Xanh d∆∞∆°ng', 'Xanh l√°', 'Xanh ƒëen']
        
        # T√¨m c√°c m√†u c·ª• th·ªÉ tr∆∞·ªõc
        for color in priority_colors:
            for keyword in color_keywords.get(color, []):
                if keyword in text_lower:
                    return color
        
        # Sau ƒë√≥ t√¨m c√°c m√†u ƒë∆°n gi·∫£n
        for color, keywords in color_keywords.items():
            for keyword in keywords:
                # T√¨m t·ª´ kh√≥a trong text, ƒë·∫£m b·∫£o l√† t·ª´ ƒë·∫ßy ƒë·ªß (kh√¥ng ph·∫£i m·ªôt ph·∫ßn c·ªßa t·ª´ kh√°c)
                pattern = r'\b' + re.escape(keyword) + r'\b'
                if re.search(pattern, text_lower):
                    return color
        
        return None
    
    # 1. Ki·ªÉm tra s·ªë l∆∞·ª£ng null tr∆∞·ªõc khi fill
    null_before = df_merged['color'].isnull().sum()
    print(f"\nüìä Tr∆∞·ªõc khi fill:")
    print(f"   - S·ªë d√≤ng null: {null_before}")
    print(f"   - Theo ngu·ªìn:")
    print(df_merged[df_merged['color'].isnull()]['source'].value_counts())
    
    # 2. Extract m√†u t·ª´ description cho c√°c d√≤ng null
    print(f"\nüîÑ ƒêang extract m√†u t·ª´ description...")
    
    null_mask = df_merged['color'].isnull()
    null_rows = df_merged[null_mask]
    
    extracted_count = 0
    failed_count = 0
    
    for idx in null_rows.index:
        row = df_merged.loc[idx]
        description = row.get('description', '')
        title = row.get('title', '')
        
        # ∆Øu ti√™n t√¨m trong description, sau ƒë√≥ m·ªõi t√¨m trong title
        extracted_color = None
        
        if pd.notna(description) and str(description).strip():
            extracted_color = extract_color_from_text(description)
        
        # N·∫øu kh√¥ng t√¨m th·∫•y trong description, th·ª≠ t√¨m trong title
        if not extracted_color and pd.notna(title) and str(title).strip():
            extracted_color = extract_color_from_text(title)
        
        if extracted_color:
            df_merged.loc[idx, 'color'] = extracted_color
            extracted_count += 1
        else:
            failed_count += 1
    
    # 3. K·∫øt qu·∫£ sau khi extract
    null_after = df_merged['color'].isnull().sum()
    
    print(f"\n‚úÖ K·∫øt qu·∫£:")
    print(f"   - ƒê√£ extract ƒë∆∞·ª£c: {extracted_count} d√≤ng")
    print(f"   - Kh√¥ng extract ƒë∆∞·ª£c: {failed_count} d√≤ng")
    print(f"   - Null tr∆∞·ªõc: {null_before}")
    print(f"   - Null sau: {null_after}")
    print(f"   - ƒê√£ gi·∫£m: {null_before - null_after} d√≤ng null")
    
    # 4. Xem m·ªôt s·ªë v√≠ d·ª• ƒë√£ extract
    if extracted_count > 0:
        print(f"\nüìã M·ªôt s·ªë v√≠ d·ª• ƒë√£ extract:")
        # L·∫•y c√°c d√≤ng v·ª´a ƒë∆∞·ª£c fill (c√≥ th·ªÉ ki·ªÉm tra b·∫±ng c√°ch xem c√°c d√≤ng c√≥ color m·ªõi)
        examples = df_merged[df_merged['color'].notna()].tail(extracted_count).head(10)
        for idx, row in examples.iterrows():
            desc_preview = str(row.get('description', ''))[:60] if pd.notna(row.get('description')) else 'N/A'
            print(f"   ‚Ä¢ {row['model']}: {desc_preview}...")
            print(f"     ‚Üí M√†u: {row['color']}")
    
    # 5. Xem c√°c d√≤ng kh√¥ng extract ƒë∆∞·ª£c
    if failed_count > 0:
        print(f"\n‚ö†Ô∏è C√°c d√≤ng kh√¥ng extract ƒë∆∞·ª£c ({failed_count} d√≤ng):")
        still_null = df_merged[df_merged['color'].isnull()]
        examples = still_null[['make', 'model', 'description']].head(10)
        for idx, row in examples.iterrows():
            desc = str(row.get('description', ''))[:80] if pd.notna(row.get('description')) else 'N/A'
            print(f"   ‚Ä¢ {row['model']}: {desc}...")
    
    # 6. Th·ªëng k√™ sau khi fill
    print(f"\nüìä Th·ªëng k√™ sau khi fill:")
    print(f"   - T·ªïng s·ªë d√≤ng: {len(df_merged)}")
    print(f"   - C√≥ m√†u: {df_merged['color'].notna().sum()} ({df_merged['color'].notna().sum()/len(df_merged)*100:.1f}%)")
    print(f"   - Null: {df_merged['color'].isnull().sum()} ({df_merged['color'].isnull().sum()/len(df_merged)*100:.1f}%)")
    
    print(f"\n   Theo ngu·ªìn:")
    for source in df_merged['source'].unique():
        source_df = df_merged[df_merged['source'] == source]
        null_source = source_df['color'].isnull().sum()
        print(f"   - {source}: {null_source}/{len(source_df)} null ({null_source/len(source_df)*100:.1f}%)")
    
    # 7. Ph√¢n b·ªë m√†u sau khi fill
    print(f"\nüìä Ph√¢n b·ªë m√†u sau khi fill:")
    print(df_merged['color'].value_counts().head(15))
```

```python
# Merge l·∫°i file invalid_versions_data ƒë√£ s·ª≠a v√†o df_merged
import pandas as pd
from pathlib import Path

if 'df_merged' not in globals():
    raise NameError("df_merged kh√¥ng t·ªìn t·∫°i!")

# Backup df_merged tr∆∞·ªõc khi merge
from datetime import datetime
backup_file = Path(f"df_merged_backup_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv")
df_merged.to_csv(backup_file, index=False, encoding='utf-8-sig')
print(f"‚úÖ ƒê√£ backup df_merged v√†o: {backup_file}")
print(f"   - S·ªë d√≤ng backup: {len(df_merged)}")

# Load file ƒë√£ s·ª≠a
fixed_file = Path("invalid_versions_data.csv")
if not fixed_file.exists():
    raise FileNotFoundError(f"Kh√¥ng t√¨m th·∫•y file {fixed_file}")

df_fixed = pd.read_csv(fixed_file, encoding='utf-8-sig')
print(f"\nüìÅ ƒê√£ load file ƒë√£ s·ª≠a: {fixed_file}")
print(f"   - S·ªë d√≤ng: {len(df_fixed)}")

# Ki·ªÉm tra c√°c c·ªôt c·∫ßn thi·∫øt
required_cols = ['row_index', 'version']
missing_cols = [col for col in required_cols if col not in df_fixed.columns]
if missing_cols:
    raise ValueError(f"Thi·∫øu c√°c c·ªôt: {missing_cols}")

# Reset index c·ªßa df_merged ƒë·ªÉ match v·ªõi row_index
df_merged_reset = df_merged.reset_index(drop=True)
max_index = df_merged_reset.index.max()

print(f"\nüìä Th√¥ng tin tr∆∞·ªõc khi merge:")
print(f"   - S·ªë d√≤ng trong df_merged: {len(df_merged_reset)}")
print(f"   - Index max: {max_index}")

# Ki·ªÉm tra row_index c√≥ h·ª£p l·ªá kh√¥ng
invalid_indices = df_fixed[df_fixed['row_index'] > max_index]
if len(invalid_indices) > 0:
    print(f"\n‚ö†Ô∏è C·∫£nh b√°o: C√≥ {len(invalid_indices)} row_index kh√¥ng h·ª£p l·ªá (l·ªõn h∆°n {max_index})")
    print(f"   C√°c row_index kh√¥ng h·ª£p l·ªá: {sorted(invalid_indices['row_index'].unique().tolist())[:10]}")
    # Lo·∫°i b·ªè c√°c row_index kh√¥ng h·ª£p l·ªá
    df_fixed = df_fixed[df_fixed['row_index'] <= max_index]
    print(f"   ƒê√£ lo·∫°i b·ªè, c√≤n l·∫°i {len(df_fixed)} d√≤ng h·ª£p l·ªá")

# Merge: Update c√°c c·ªôt t·ª´ df_fixed v√†o df_merged
print(f"\nüîÑ ƒêang merge l·∫°i v√†o df_merged...")

# L·∫•y t·∫•t c·∫£ c√°c c·ªôt t·ª´ df_fixed (tr·ª´ row_index v√† version_sai)
cols_to_update = [col for col in df_fixed.columns 
                  if col not in ['row_index', 'version_sai']]

updated_count = 0
updated_rows = []

for idx, row in df_fixed.iterrows():
    row_idx = int(row['row_index'])
    
    if row_idx <= max_index:
        # Update t·∫•t c·∫£ c√°c c·ªôt t·ª´ df_fixed v√†o df_merged
        for col in cols_to_update:
            if col in df_merged_reset.columns:
                old_value = df_merged_reset.loc[row_idx, col]
                new_value = row[col]
                
                # Ch·ªâ update n·∫øu kh√°c nhau (ƒë·∫∑c bi·ªát l√† version)
                if pd.isna(old_value) and pd.isna(new_value):
                    continue
                elif pd.isna(old_value) or pd.isna(new_value) or str(old_value) != str(new_value):
                    df_merged_reset.loc[row_idx, col] = new_value
                    if col == 'version':  # Track version changes
                        updated_rows.append({
                            'row_index': row_idx,
                            'model': row.get('model', ''),
                            'year': row.get('year', ''),
                            'old_version': old_value,
                            'new_version': new_value
                        })
        
        updated_count += 1

print(f"‚úÖ ƒê√£ update {updated_count} d√≤ng trong df_merged")

# G√°n l·∫°i v√†o df_merged
df_merged = df_merged_reset.copy()

# Verify: Ki·ªÉm tra l·∫°i
print(f"\nüîç ƒêang verify...")
verification_errors = []

for idx, row in df_fixed.iterrows():
    row_idx = int(row['row_index'])
    if row_idx <= max_index:
        # Ki·ªÉm tra version ƒë√£ ƒë∆∞·ª£c update ch∆∞a
        current_version = df_merged.loc[row_idx, 'version']
        expected_version = row['version']
        
        if str(current_version) != str(expected_version):
            verification_errors.append({
                'row_index': row_idx,
                'model': row.get('model', ''),
                'year': row.get('year', ''),
                'expected': expected_version,
                'actual': current_version
            })

if verification_errors:
    print(f"‚ö†Ô∏è C√≥ {len(verification_errors)} d√≤ng ch∆∞a ƒë∆∞·ª£c update ƒë√∫ng:")
    for error in verification_errors[:10]:
        print(f"   - Row {error['row_index']} ({error['model']} {error['year']}): Expected '{error['expected']}', got '{error['actual']}'")
else:
    print(f"‚úÖ T·∫•t c·∫£ ƒë√£ ƒë∆∞·ª£c merge ƒë√∫ng!")

# In th√¥ng tin df m·ªõi ƒë·ªÉ ki·ªÉm tra
print(f"\n" + "="*80)
print("TH√îNG TIN DF_MERGED SAU KHI MERGE")
print("="*80)

print(f"\nüìä T·ªïng quan:")
print(f"   - T·ªïng s·ªë d√≤ng: {len(df_merged)}")
print(f"   - S·ªë c·ªôt: {len(df_merged.columns)}")
print(f"   - C√°c c·ªôt: {list(df_merged.columns)}")

print(f"\nüìä Th·ªëng k√™ theo model:")
model_counts = df_merged['model'].value_counts()
print(model_counts.head(10))

print(f"\nüìä Th·ªëng k√™ version ƒë√£ s·ª≠a (m·∫´u):")
if updated_rows:
    df_updated = pd.DataFrame(updated_rows)
    print(f"   - T·ªïng s·ªë version ƒë√£ s·ª≠a: {len(df_updated)}")
    print(f"\n   M·ªôt s·ªë version ƒë√£ s·ª≠a:")
    for _, row in df_updated.head(20).iterrows():
        print(f"      Row {row['row_index']}: {row['model']} {row['year']}")
        print(f"         '{row['old_version']}' -> '{row['new_version']}'")

# Ki·ªÉm tra c√°c version sai c√≤n l·∫°i kh√¥ng
print(f"\nüîç Ki·ªÉm tra c√°c version sai c√≤n l·∫°i:")
remaining_invalid = []

# L·∫•y danh s√°ch version sai t·ª´ file g·ªëc (n·∫øu c√≥ c·ªôt version_sai)
if 'version_sai' in df_fixed.columns:
    invalid_versions_list = df_fixed['version_sai'].dropna().unique().tolist()
    
    for version_sai in invalid_versions_list[:20]:  # Ki·ªÉm tra 20 ƒë·∫ßu ti√™n
        count = len(df_merged[df_merged['version'] == version_sai])
        if count > 0:
            remaining_invalid.append((version_sai, count))

if remaining_invalid:
    print(f"   ‚ö†Ô∏è V·∫´n c√≤n m·ªôt s·ªë version sai:")
    for version, count in remaining_invalid[:10]:
        print(f"      - '{version}': {count} d√≤ng")
else:
    print(f"   ‚úÖ Kh√¥ng c√≤n version sai n√†o trong danh s√°ch!")

# Hi·ªÉn th·ªã m·ªôt s·ªë d√≤ng m·∫´u
print(f"\nüìã M·ªôt s·ªë d√≤ng m·∫´u sau khi merge:")
display(df_merged[['make', 'model', 'year', 'version', 'color', 'price_vnd']].head(20))

# Th·ªëng k√™ version theo model v√† year (m·∫´u)
print(f"\nüìä Th·ªëng k√™ version theo model v√† year (m·∫´u - 5 model ƒë·∫ßu):")
for model in df_merged['model'].value_counts().head(5).index:
    model_data = df_merged[df_merged['model'] == model]
    print(f"\n   {model}:")
    for year in sorted(model_data['year'].dropna().unique())[:3]:  # 3 nƒÉm ƒë·∫ßu
        year_data = model_data[model_data['year'] == year]
        versions = sorted(year_data['version'].dropna().unique().tolist())
        print(f"      {int(year)}: {len(versions)} versions - {versions[:5]}{'...' if len(versions) > 5 else ''}")

print(f"\n‚úÖ Ho√†n th√†nh merge!")
print(f"   - Backup file: {backup_file}")
print(f"   - S·ªë d√≤ng ƒë√£ update: {updated_count}")
print(f"   - S·ªë d√≤ng trong df_merged: {len(df_merged)}")
```

```python
# Train nhi·ªÅu model - phi√™n b·∫£n ƒë√£ s·ª≠a l·ªói
print("ü§ñ TRAIN NHI·ªÄU MODEL ƒê·ªÇ D·ª∞ ƒêO√ÅN GI√Å XE (ƒê√É S·ª¨A)")
print("\n" + "="*60)

if 'df_merged' in locals():
    import pandas as pd
    import numpy as np
    from sklearn.model_selection import train_test_split
    from sklearn.preprocessing import LabelEncoder
    from sklearn.linear_model import LinearRegression, Ridge, Lasso
    from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
    import warnings
    warnings.filterwarnings('ignore')
    
    # Ki·ªÉm tra XGBoost
    try:
        import xgboost as xgb
        XGBOOST_AVAILABLE = True
    except ImportError:
        XGBOOST_AVAILABLE = False
    
    # 1. Chu·∫©n b·ªã d·ªØ li·ªáu
    print("\n1Ô∏è‚É£ CHU·∫®N B·ªä D·ªÆ LI·ªÜU:")
    
    feature_cols = ['make', 'model', 'year', 'version', 'color', 'mileage']
    target_col = 'price_vnd'
    
    df_train = df_merged[feature_cols + [target_col]].copy()
    
    # Lo·∫°i b·ªè null
    print(f"   - Tr∆∞·ªõc khi x·ª≠ l√Ω: {len(df_train)} d√≤ng")
    df_train = df_train.dropna()
    print(f"   - Sau khi lo·∫°i b·ªè null: {len(df_train)} d√≤ng")
    
    # Ki·ªÉm tra v√† x·ª≠ l√Ω infinity
    print(f"\n   Ki·ªÉm tra infinity v√† gi√° tr·ªã b·∫•t th∆∞·ªùng...")
    for col in feature_cols + [target_col]:
        if df_train[col].dtype in [np.float64, np.float32, np.int64, np.int32]:
            inf_count = np.isinf(df_train[col]).sum()
            if inf_count > 0:
                print(f"     ‚ö†Ô∏è {col}: {inf_count} gi√° tr·ªã infinity")
                df_train = df_train[~np.isinf(df_train[col])]
    
    # Ki·ªÉm tra gi√° tr·ªã qu√° l·ªõn
    for col in feature_cols + [target_col]:
        if df_train[col].dtype in [np.float64, np.float32]:
            max_val = df_train[col].max()
            if max_val > 1e10:  # Gi√° tr·ªã qu√° l·ªõn
                print(f"     ‚ö†Ô∏è {col}: c√≥ gi√° tr·ªã l·ªõn nh·∫•t = {max_val}")
    
    print(f"   - Sau khi x·ª≠ l√Ω: {len(df_train)} d√≤ng")
    
    # 2. Feature Engineering
    print(f"\n2Ô∏è‚É£ FEATURE ENGINEERING:")
    
    label_encoders = {}
    X_encoded = pd.DataFrame()
    
    for col in feature_cols:
        if df_train[col].dtype == 'object' or df_train[col].dtype.name == 'category':
            le = LabelEncoder()
            X_encoded[col] = le.fit_transform(df_train[col].astype(str))
            label_encoders[col] = le
            print(f"   - {col}: Label encoded ({df_train[col].nunique()} categories)")
        else:
            X_encoded[col] = df_train[col]
            print(f"   - {col}: Gi·ªØ nguy√™n (numerical)")
    
    y = df_train[target_col].values
    
    # Ki·ªÉm tra l·∫°i NaN/infinity sau khi encode
    print(f"\n   Ki·ªÉm tra NaN/infinity sau khi encode...")
    if X_encoded.isnull().sum().sum() > 0:
        print(f"     ‚ö†Ô∏è C√≥ NaN trong X_encoded")
        X_encoded = X_encoded.fillna(X_encoded.median())
    
    if np.isinf(X_encoded.values).sum() > 0:
        print(f"     ‚ö†Ô∏è C√≥ infinity trong X_encoded")
        X_encoded = X_encoded.replace([np.inf, -np.inf], np.nan)
        X_encoded = X_encoded.fillna(X_encoded.median())
    
    if np.isnan(y).sum() > 0:
        print(f"     ‚ö†Ô∏è C√≥ NaN trong y")
        mask = ~np.isnan(y)
        X_encoded = X_encoded[mask]
        y = y[mask]
    
    if np.isinf(y).sum() > 0:
        print(f"     ‚ö†Ô∏è C√≥ infinity trong y")
        mask = ~np.isinf(y)
        X_encoded = X_encoded[mask]
        y = y[mask]
    
    print(f"   - Sau khi x·ª≠ l√Ω: {len(X_encoded)} d√≤ng")
    
    # 3. Chia train/test
    print(f"\n3Ô∏è‚É£ CHIA TRAIN/TEST:")
    X_train, X_test, y_train, y_test = train_test_split(
        X_encoded, y, test_size=0.2, random_state=42
    )
    
    print(f"   - Train: {len(X_train)} d√≤ng")
    print(f"   - Test: {len(X_test)} d√≤ng")
    
    # Reset index ƒë·ªÉ tr√°nh l·ªói
    X_train = X_train.reset_index(drop=True)
    X_test = X_test.reset_index(drop=True)
    df_train_reset = df_train.reset_index(drop=True)
    
    # 4. Train c√°c model
    print(f"\n4Ô∏è‚É£ TRAIN C√ÅC MODEL:")
    
    models = {
        'Linear Regression': LinearRegression(),
        'Ridge Regression': Ridge(alpha=1.0),
        'Lasso Regression': Lasso(alpha=1.0),
        'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1),
        'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
    }
    
    if XGBOOST_AVAILABLE:
        models['XGBoost'] = xgb.XGBRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    
    results = {}
    
    for name, model in models.items():
        print(f"\n   üîπ Training {name}...")
        try:
            model.fit(X_train, y_train)
            
            y_pred_train = model.predict(X_train)
            y_pred_test = model.predict(X_test)
            
            train_mae = mean_absolute_error(y_train, y_pred_train)
            test_mae = mean_absolute_error(y_test, y_pred_test)
            train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
            test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
            train_r2 = r2_score(y_train, y_pred_train)
            test_r2 = r2_score(y_test, y_pred_test)
            
            results[name] = {
                'model': model,
                'train_mae': train_mae,
                'test_mae': test_mae,
                'train_rmse': train_rmse,
                'test_rmse': test_rmse,
                'train_r2': train_r2,
                'test_r2': test_r2
            }
            
            print(f"      ‚úÖ Ho√†n th√†nh")
            print(f"         Train MAE: {train_mae:.0f} tri·ªáu")
            print(f"         Test MAE: {test_mae:.0f} tri·ªáu")
            print(f"         Test R¬≤: {test_r2:.3f}")
            
        except Exception as e:
            print(f"      ‚ùå L·ªói: {str(e)[:100]}")
    
    # 5. So s√°nh k·∫øt qu·∫£
    print(f"\n5Ô∏è‚É£ SO S√ÅNH K·∫æT QU·∫¢:")
    
    comparison_df = pd.DataFrame({
        'Model': list(results.keys()),
        'Train MAE': [results[m]['train_mae'] for m in results.keys()],
        'Test MAE': [results[m]['test_mae'] for m in results.keys()],
        'Train RMSE': [results[m]['train_rmse'] for m in results.keys()],
        'Test RMSE': [results[m]['test_rmse'] for m in results.keys()],
        'Train R¬≤': [results[m]['train_r2'] for m in results.keys()],
        'Test R¬≤': [results[m]['test_r2'] for m in results.keys()]
    })
    
    comparison_df = comparison_df.sort_values('Test MAE')
    
    print("\n   üìä B·∫£ng so s√°nh:")
    print(comparison_df.to_string(index=False))
    
    # 6. Model t·ªët nh·∫•t
    print(f"\n6Ô∏è‚É£ MODEL T·ªêT NH·∫§T:")
    best_model_name = comparison_df.iloc[0]['Model']
    best_model = results[best_model_name]['model']
    
    print(f"   ‚úÖ {best_model_name}")
    print(f"   - Test MAE: {results[best_model_name]['test_mae']:.0f} tri·ªáu VND")
    print(f"   - Test RMSE: {results[best_model_name]['test_rmse']:.0f} tri·ªáu VND")
    print(f"   - Test R¬≤: {results[best_model_name]['test_r2']:.3f}")
    
    # 7. Feature importance
    print(f"\n7Ô∏è‚É£ FEATURE IMPORTANCE:")
    
    if hasattr(best_model, 'feature_importances_'):
        feature_importance = pd.DataFrame({
            'Feature': feature_cols,
            'Importance': best_model.feature_importances_
        }).sort_values('Importance', ascending=False)
        
        print(f"\n   {best_model_name}:")
        print(feature_importance.to_string(index=False))
    
    # 8. V√≠ d·ª• d·ª± ƒëo√°n (ƒë√£ s·ª≠a)
    print(f"\n8Ô∏è‚É£ V√ç D·ª§ D·ª∞ ƒêO√ÅN:")
    
    # L·∫•y indices t·ª´ test set (ƒë√£ reset index)
    test_indices = X_test.index[:5]
    
    # L·∫•y indices t∆∞∆°ng ·ª©ng trong df_train_reset
    # V√¨ ƒë√£ reset index, c·∫ßn map l·∫°i
    train_indices = []
    for i, test_idx in enumerate(test_indices):
        # T√¨m index trong df_train_reset t∆∞∆°ng ·ª©ng v·ªõi test sample th·ª© i
        # S·ª≠ d·ª•ng iloc thay v√¨ loc
        train_indices.append(len(X_train) + test_idx)
    
    for i, test_idx in enumerate(test_indices[:5]):
        actual = y_test[test_idx]
        predicted = best_model.predict(X_test.iloc[[test_idx]])[0]
        error = abs(actual - predicted)
        error_pct = (error / actual) * 100 if actual > 0 else 0
        
        # L·∫•y th√¥ng tin t·ª´ df_train_reset
        original_idx = len(X_train) + test_idx
        if original_idx < len(df_train_reset):
            original_row = df_train_reset.iloc[original_idx]
            
            print(f"\n   V√≠ d·ª• {i+1}:")
            print(f"   - Make: {original_row['make']}")
            print(f"   - Model: {original_row['model']}")
            print(f"   - Year: {original_row['year']}")
            print(f"   - Version: {original_row['version']}")
            print(f"   - Color: {original_row['color']}")
            print(f"   - Mileage: {original_row['mileage']}")
            print(f"   - Gi√° th·ª±c t·∫ø: {actual:.0f} tri·ªáu")
            print(f"   - Gi√° d·ª± ƒëo√°n: {predicted:.0f} tri·ªáu")
            print(f"   - Sai s·ªë: {error:.0f} tri·ªáu ({error_pct:.1f}%)")
```
